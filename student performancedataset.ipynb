# Data Visualization
Scatter Plot – Study Hours vs Exam Score

Data Visualization is the process of representing data using graphs, charts, and plots to make the information easier to understand. It helps identify patterns, trends, and relationships in the dataset.

In the Student Performance project, visualization techniques such as scatter plots, bar charts, box plots, and heatmaps are used to analyze how factors like study hours, attendance, sleep hours, and motivation level affect exam scores. Data visualization helps in gaining clear insights and making better decisions based on the data.

```python
plt.scatter(df["Hours_Studied"], df["Exam_Score"])
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()
```

# Bar Plot – Motivation Level vs Score

A bar plot is used to compare values between different categories. In this analysis, the bar plot shows the relationship between motivation level and exam score.

```python
sns.barplot(x="Motivation_Level", y="Exam_Score", data=df)
plt.title("Motivation Level vs Exam Score")
plt.show()
```

# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing and understanding a dataset before applying any machine learning models. It helps in identifying patterns, trends, and relationships between variables in the data. EDA also helps detect missing values, outliers, and data distribution.

In the Student Performance dataset, EDA is used to analyze factors such as study hours, attendance, sleep hours, and motivation level to understand how they affect students' exam scores. Various visualization techniques like scatter plots, bar charts, and box plots are used to explore the relationships between variables.

EDA helps researchers and educators gain meaningful insights from the data and supports better decision-making and prediction of student performance.

```python
plt.figure(figsize=(8,5))
plt.hist(df['Exam_Score'], bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot – Exam Score by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='Exam_Score', data=df, palette='Set2')
plt.title("Exam Score by Gender")
plt.show()
```

# Pie Chart

```python
#Pie Chart – Gender Distribution
df['Gender'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['skyblue','pink'],
    figsize=(6,6)
)

plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
#Line Plot – Hours Studied vs Score Trend
df_sorted = df.sort_values(by='Hours_Studied')

plt.figure(figsize=(8,5))
plt.plot(df_sorted['Hours_Studied'], df_sorted['Exam_Score'], color='blue')
plt.title("Study Hours vs Score Trend")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

# Heatmap

A heatmap is a data visualization technique used to show the correlation between different variables in a dataset. It uses different colors to represent the strength of relationships between variables.

In the Student Performance dataset, a heatmap helps analyze how features like study hours, attendance, sleep hours, and motivation level are related to exam scores. Strong correlations appear in darker colors, while weak correlations appear in lighter colors.

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
